# Visualise

```python
def pairwiseComperison(data):
    
    idx = 7
    models = ['MadGan', 'AutoEncoder', 'VanLstm', 'IsoForest', 'GMM', 'OcSVM', 'BiLstmGan', 'CnnGan', 'MTSAaGAN']
    data = data.transpose()
    tm = np.array([row[8] for row in data])
    bm = np.array([row[idx] for row in data])
    #bm = np.array([max(row[1:]) for row in data])
    
    f, ax = plt.subplots(figsize=(4, 4))
    ax.set_xlim(right=1)
    ax.set_ylim(top=1)
    ax.fill_between([0, 1], [0, 1], transform=ax.transAxes, alpha=.3)
    ax.scatter(tm, bm)
    ax.plot([0, 1], [0, 1], transform=ax.transAxes)
    plt.xlabel('ALGAN', fontsize=16)
    plt.ylabel(models[idx], fontsize=16)
    plt.title(title, fontsize=16)

pairwiseComperison(data)
```
