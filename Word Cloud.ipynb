# Creating Word Cloud From Text Data

### Import required libraries

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

### Creat Word Cloud

```python
sentences = ['I like to paly in the afternoon', 
             'She likes to sing in the morning',
            'What a wonderful day it is',
            'Data science is fun',
            'Election is comming soon',
            'They are trying to buy some instrument',
            'Data Science is going to be more useful in the coming days',
            'I am interested in data science']

text = ' '.join(sentences).replace('hashtag', '').replace('rt','').replace('user', '').replace('url', '').replace('repeated', '').replace('will', '').replace('allcaps', '')

# lower max_font_size, change the maximum number of word and lighten the background:
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
